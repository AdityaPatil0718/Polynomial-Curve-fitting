```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setting seed as 20
np.random.seed(20)
```

```python
# X_coordinate
x = np.linspace(-20,20,10)
```

```python
#polynomial equation
y=2*x+5
```

```python
#storing coefficient by using polyfit
coeff = np.polyfit(x,y,2)
x_new = np.linspace(-20,20,100)
y_new = np.poly1d(coeff)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'o')
plt.subplot(2,1,2)
plt.plot(x,y,'o', x_new,y_new(x_new),'-')
```
